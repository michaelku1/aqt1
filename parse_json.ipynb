```python
list_of_classes = [3,5,8]
```

```python
# retrieve image ids corresponding to the category train
# ids = []
# annotations = []
# images = []
dict_all = {}

# only need the lists stored in dict
for cls in list_of_classes:
    # store image id, gorup of annotations, this is to store annotation of the specified categories
    dict_all[cls] = {'image_id': [], 'annotations':[]}
```

```python
dict_all
```

```python
# define classes to be added
categories = []
for cls_idx, cls in enumerate(list_of_classes):
    categories.append({'id': cls_idx + 1, 'name': file['categories'][cls-1]['name']})
#     import pdb; pdb.set_trace()
#     file['categories'] = [{'id': cls_idx, 'name': file['categories'][cls_idx]['name']}]
```

```python
for i in range(len(file['annotations'])):
    for cls_idx, cls in enumerate(list_of_classes):
        if file['annotations'][i]['category_id']==cls:
            file['annotations'][i]['category_id'] = cls_idx + 1 # relabel
            dict_all[cls]['image_id'].append(file['annotations'][i]['image_id'])
            dict_all[cls]['annotations'].append(file['annotations'][i])           
# ids = set(ids) # need to check if this is needed
```

```python
for category_id in list(dict_all.keys()):
    image_set = list(set(dict_all[category_id]['image_id'])) # all image ids
    dict_all[category_id]['image_id'] = image_set
```

```python
import numpy as np

# get all image ids
all_image_ids = []
for category_id in list(dict_all.keys()):
    all_image_ids.append(dict_all[category_id]['image_id'])

# need to get rid of overlaps for image ids
all_image_ids = list(set(np.concatenate(all_image_ids)))

# get all subsets of images corresponding to the specified categories 
list_of_images = []
annotations = []

for image in file['images']:
    for image_id in all_image_ids:
        if image_id==image['id']:
            list_of_images.append(image)
            
# get all annotations (note that there is no overlap with annotations)
for category_id in list(dict_all.keys()):
    annotations.append(dict_all[category_id]['annotations'])
```

```python
# check number of data points for each category
data_num = []
for cat in list_of_classes:
    data_num.append(len(dict_all[cat]['image_id']))
```

```python
new_json = {}
new_json['images'] = list_of_images
new_json['annotations'] = list(np.concatenate(annotations))
new_json['categories'] = categories
```

## parse images for increasing fog density

```python
import glob
import shutil
import os
import pathlib as path
```

```python
path.posixpath('/scratch2/users/cku/adaptation/datasets/cityscapes')
```

```python
train = '/scratch2/users/cku/adaptation/datasets/cityscapes/leftImg8bit_foggy/train'
val = '/scratch2/users/cku/adaptation/datasets/cityscapes/leftImg8bit_foggy/val'
list_of_train_folder_names = []
list_of_val_folder_names = []
folders_train = glob.glob(train+'/*')
folders_val = glob.glob(val+'/*')
for folder in folders_train:
    name = folder.split('/')[-1]
    list_of_train_folder_names.append(name)
for folder in folders_val:
    name = folder.split('/')[-1]
    list_of_val_folder_names.append(name)
```

```python
list_of_train_folder_names
```

```python
list_of_val_folder_names
```

```python
root = '/scratch2/users/cku/adaptation/datasets/cityscapes'
foggy_folder = '/scratch2/users/cku/adaptation/datasets/cityscapes/leftImg8bit_foggy'

folders = ['leftImg8bit_foggy_0.005', 'leftImg8bit_foggy_0.01', 'leftImg8bit_foggy_0.02']

# mkdir for the three folders, under which there is a train and val
for folder in folders:
#     density = folder.split('_')[-1]
    folder_path_ = root + '/' + folder
    if not os.path.isdir(folder_path_):
        os.makedirs(folder_path_, exist_ok = True)
    else:
        train = folder_path_ + '/' + 'train'
        val = folder_path_ + '/' + 'val'
        if not os.path.isdir(train):
            os.makedirs(train, exist_ok=True)
        if not os.path.isdir(val):
            os.makedirs(val, exist_ok=True)
```

```python
sub_folders = ['train', 'val']
for folder in folders:
    folder_path_ = root + '/' + folder
    for subfolder in sub_folders:
        if subfolder == 'train':
            new_folder_path_ = folder_path_ + '/' + subfolder
            for name in list_of_train_folder_names:
                full_path_name = new_folder_path_ + '/' + name
#                 import pdb; pdb.set_trace()
                if not os.path.isdir(full_path_name):
                    os.makedirs(full_path_name, exist_ok = True)
        elif subfolder =='val':
            new_folder_path_ = folder_path_ + '/' + subfolder
            for name in list_of_val_folder_names:
                full_path_name = new_folder_path_ + '/' + name
                if not os.path.isdir(full_path_name):
                    os.makedirs(full_path_name, exist_ok = True)
```

```python
# copy files to the corresponding folder
for each in folders:
    foggy_folder_made = root+ '/' + each # the made folder
    density = each.split('_')[-1] # density
    for city_name in list_of_train_folder_names:
        full_path = foggy_folder + '/' + 'train' + '/' + city_name # each city src path
        list_of_imgs = glob.glob(full_path + f'/*{density}*')
        for img_path in list_of_imgs:
#             import pdb; pdb.set_trace()
            suffix = img_path.split('/')[-1]
            shutil.copyfile(img_path, foggy_folder_made + '/' + 'train' + '/' + city_name + '/' + suffix)
            
    for city_name in list_of_val_folder_names:
        full_path = foggy_folder + '/' + 'val' + '/' + city_name
        list_of_imgs = glob.glob(full_path + f'/*{density}*')
        for img_path in list_of_imgs:
            suffix = img_path.split('/')[-1]
            shutil.copyfile(img_path, foggy_folder_made + '/' + 'val' + '/' + city_name + '/' + suffix)
```
